#Task 3
Perform the Below Visualizations.

#Univariate

```python
sns.distplot(df.living_area_renov)
```

```python
plt.pie(df['Number of schools nearby'].value_counts(),[0,0.1,0.1],labels=['2','1','3'],autopct='%1.1f%%',shadow=True,colors=['blue','yellow','green'])
plt.title('Number of schools nearby')
plt.show()
```

```python
sns.barplot(x= df['Number of schools nearby'].value_counts().index, y = df['Number of schools nearby'].value_counts())
```

#Bivariate

```python
sns.lineplot(x=df['Number of schools nearby'],y=df['Built Year'])
```

```python
sns.scatterplot(x=df['Number of schools nearby'],y=df['Built Year'])
```

#Multivariate

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```
